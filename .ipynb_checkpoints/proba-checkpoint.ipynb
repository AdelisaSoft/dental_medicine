```python
import pandas as pd
import sklearn
```

```python
df = pd.read_excel("test_podaci_aida.xlsx")
```

```python
df.describe()
```

```python
df.head()
```

```python
def map(x):
    if x == "Da":
        return 1 
    return 0

def korekcijaRizika(x):
    if x[0] == 'v':
        return "visoki"
    elif x[0] == 's':
        return "srednji"
    return "niski"
    
    
brojac = 1
rename_dict = {}
for col in df.columns:
    if col != "RIZIK":
        rename_dict[col] = "Q"+str(brojac)
        brojac += 1
df = df.rename(columns=rename_dict)
```

```python
df.head()
```

```python
df = df.dropna(subset=['RIZIK'])
```

```python
df['RIZIK'] = df['RIZIK'].apply(lambda x: korekcijaRizika(x))
df['Q3'] = df['Q3'].apply(lambda x: map(x))
df['Q4'] = df['Q4'].apply(lambda x: map(x))
df['Q5'] = df['Q5'].apply(lambda x: map(x))
df['Q6'] = df['Q6'].apply(lambda x: map(x))
df['Q7'] = df['Q7'].apply(lambda x: map(x))
df['Q8'] = df['Q8'].apply(lambda x: map(x))
```

```python
df.head()
```

```python
one_hot = pd.get_dummies(df['Q1'])
df = df.drop('Q1',axis = 1)
df = df.join(one_hot)

one_hot = pd.get_dummies(df['Q2'])
df = df.drop('Q2',axis = 1)
df = df.join(one_hot)

one_hot = pd.get_dummies(df['Q9'])
df = df.drop('Q9',axis = 1)
df = df.join(one_hot)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
clf = LogisticRegression(random_state=0, multi_class="multinomial")
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```
